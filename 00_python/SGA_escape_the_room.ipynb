The game name is ESCAPE THE ROOM.
It's a well-known family of quest games. The idea is that main character finds himself (wakes up) in a strange room and needs to guess a way out of it. To complete the game the player needs to look at all objects in the room and try to figure out how and what to take/use/open in order to succeed. The only win scenario is opening the room dor. The player loses if gives up so the main character remains in the room.
The game implements following actions: "look around" to look around in the room, "move" to move across the room, "inventory" to show collected items. Some action require object name thay are applied to: "look at [object]" to look at objects in the room or in the inventory, "take [object]" to take an object, "open [object]", "use [object]" to use an object. There are also some supportive actions: "hint" to give a hint to the player, "help" to explain how the game input is organized, "exit" to exit to the main menu.
As the latest action implies there is the main menu in the game. It is supported by folloeing actions: "new game" to start a new game, "resume" to resume the game in progress, "description" to explain the main idea of the game to the player, "quit" to quit the game.
There are also several auxillary functions to support the gameplay and avoid some redunduncy.

```python
msg_pack = [
    ["""### You wake up in a tiny room. Very quiet chilling music can be heard from a small ceiling speaker.
    The only feeling you feel is tranquillity. The only thougt is: Where am I?""",
    """### You start recalling the last night. The party and the bar…""",
    """### Yesterday you met with three frinds of yours to give a small good-buy party to one of you 
    flying to Singapore for a business trip.""",
    """### Yes, it was such a party hard… 
    Probably it was a bad idea to mix a bloody Mary and a shot of B52.""",
    """### Now you remember that you and your friends went by taxy to the airport 
    for someone of you to catch the evening flight.""",
    """### You remember the airport. 
    And that everyone of your friends forgot who exacly of you four needs to catch the flight.""",
    """### After some discussion and a fair coin tossing to choose who's flying away 
    some of you proposed to look into ticket details.""",
    """### One of your friends happily took the plane. 
    The rest three of you were very tired and decided to stay in a capsule-like hotel in the airport building.""",
    "### So you are in a hotel, this is a room. No shocking mistery. Disappointimg a bit, right?",
    "### You are just one step from meeting your friends, talk about last night and have so much fun. Go on!",
    "### CONGRATULATIONS! You have successfully completed the game!"],
    ["""This is a barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed.  
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set.""",
    """Still the same barely lit narrow rectangular room. 
    All its width allows is a bed and a small table adjacent to the head of the bed. 
    There is a TV set hanging on the opposite narrow wall so you could watch it just lying on the bed. 
    And there is a door! It's in a long wall between the foot of the bed and the TV set."""],
    ["A regular bed on four legs. It's too dark to see what's under.",
    "A regular bed on four legs. It's too dark to see what's under.",
    "A regular bed on four legs. It's too dark to see what's under.",
    "A regular bed on four legs. It's too dark to see what's under.",
    "A regular bed on four legs. It's too dark to see what's under.",
    "A regular bed on four legs. It's too dark to see what's under.",
    "A regular bed on four legs. It's too dark to see what's under.",
    "A regular bed on four legs. There's a key under the bed.",
    "A regular bed on four legs. It's too dark to see what's under \
    but you know that there's nothing interesting anymore.",
    "A regular bed on four legs. It's too dark to see what's under \
    but you know that there's nothing interesting anymore."],
    ["A tiny one-leg table. There is a TV remote control on the table.",
    "A tiny one-leg table. Nothing special.",
    "A tiny one-leg table. Nothing special.",
    "A tiny one-leg table. There is an inoperatable remote control on it. Nothing special.",
    "A tiny one-leg table. There is an inoperatable remote control on it. Nothing special.",
    "A tiny one-leg table. There is an inoperatable remote control on it. Nothing special.",
    "A tiny one-leg table. There is an inoperatable remote control on it. Nothing special.",
    "A tiny one-leg table. There is an inoperatable remote control on it. Nothing special.",
    "A tiny one-leg table. There is an inoperatable remote control on it. Nothing special.",
    "A tiny one-leg table. There is an inoperatable remote control on it. Nothing special."],
    ["Not so fast! Sudden moves can provoke headakes.",
    "The TV set is off but its red LED is shining. Oops! There's a flashlight lying on the TV set.",
    "The TV set is off but its red LED is shining. Wait! There's a flashlight lying on the TV set.",
    "The TV set is off but its red LED is shining. Wait! There's a flashlight lying on the TV set.",
    "The TV set is off.",
    "The TV set is off.",
    "The TV set is off.",
    "The TV set is off.",
    "The TV set is off.",
    "The TV set is off."],
    ["The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is a key hole in the the door knob.",
    "The door is wooden. There is the key in the the door knob."],
    ["A black plastic remote control with many buttons. Quite unusual, huh?",
    "A black plastic remote control with many buttons. Quite unusual, huh?",
    "A black plastic remote control with many buttons. Quite unusual, huh?",
    "An inoperatable remote control. Nothing special.",
    "An inoperatable remote control. Nothing special.",
    "An inoperatable remote control. Nothing special.",
    "An inoperatable remote control. Nothing special.",
    "An inoperatable remote control. Nothing special.",
    "An inoperatable remote control. Nothing special.",
    "An inoperatable remote control. Nothing special."],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "A small thin flashlight.",
    "A small thin flashlight.",
    "A small thin flashlight.",
    "A small thin flashlight.",
    "A small thin flashlight.",
    "A small thin flashlight lying under the bed.",
    "A small thin flashlight lying under the bed.",
    "A small thin flashlight lying under the bed."],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "A flat key that must perfectly fit the door knob lock.",
    "A flat key that must perfectly fit the door knob lock.",
    "You've already unlocked the door. Why waste time looking at the key?"],
    ["Please make sure you know what you are doing.",
    "Why would you open a good functioning remote? Oh, wait. Does it really function?",
    "Now you have a couple of batteries and an inoperatable remote on the table.",
    "It's already opened and useless now.",
    "It's already opened and useless now.",
    "It's already opened and useless now.",
    "It's already opened and useless now.",
    "It's already opened and useless now.",
    "It's already opened and useless now.",
    "It's already opened and useless now."],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Hmm… Looks like it needs two batteries to give some light.",
    "Why open a good operatable flashlight?",
    "Why open a good operatable flashlight?",
    "Why open a good operatable flashlight?",
    "Why open a good operatable flashlight?",
    "Why open a good operatable flashlight?"],
    ["The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "The door is locked. Now you know what Count of Monte Cristo felt.",
    "You are leaving the room. What an adventure it's been!"],
    ["Well, well… Let's see what's on TV today.",
    "You already have a remote control.",
    "You already have a remote control.",
    "The remote control doesn't work. You don't need it.",
    "The remote control doesn't work. You don't need it.",
    "The remote control doesn't work. You don't need it.",
    "The remote control doesn't work. You don't need it.",
    "The remote control doesn't work. You don't need it.",
    "The remote control doesn't work. You don't need it.",
    "The remote control doesn't work. You don't need it."],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "It doesn't work, you don't need it.",
    "Ok, here is the flashlight. Now what?",
    "No more flashlights in the room.",
    "No more flashlights in the room.",
    "No more flashlights in the room.",
    "You don't need it anymore.",
    "You don't need it anymore.",
    "You don't need it anymore."],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Got it! Now don't waste your time.",
    "No more keys to take.",
    "No more keys to take."],
    ["Please make sure you know what you are doing.",
    """The TV set turns on and starts showing some hotel promo. 
    Like 'Welcome to our capsule-like hotel' and so on. Then it suddenly turns off again.""",
    """The TV set turns on and starts showing some hotel promo.
    Like 'Welcome to our capsule-like hotel' and so on. Then it suddenly turns off again.""",
    "The remote control doesn't work.",
    "The remote control doesn't work.",
    "The remote control doesn't work.",
    "The remote control doesn't work.",
    "The remote control doesn't work.",
    "The remote control doesn't work.",
    "The remote control doesn't work."],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "You need some preparation to use the batteries.",
    "Great! Now the flashlight works. It's constantly on but it's not a problem at all.",
    "You don't have any spare batteries to use.",
    "You don't have any spare batteries to use.",
    "You don't have any spare batteries to use.",
    "You don't have any spare batteries to use."],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "No batteries, no light.",
    "No batteries, no light.",
    "You lay the flashlight under the bed. Whoa? There is a key here! Exciting, isn’t it?",
    "Whoa? There is a key under the bed! Exciting, isn’t it?",
    "What do you want to use it for?",
    "What do you want to use it for?"],
    ["Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    "Please make sure you know what you are doing.",
    """Klick-klack! Seems like the door lock is open now.
    Are you sure you want out? Maybe you wish to stay for a longer time, watch TV or something?""",
    "Already used. You gotta move on."],
    ["No items", 
    "Remote control", 
    "Remote control",
    "Batteries",
    "Batteries, flashlight",
    "Batteries, open flashlight",
    "Operatable flashlight",
    "No items",
    "Key",
    "No items"],
    ["What a nice remote control!",
    "What a nice TV set!",
    "The power source of the remote control may prove very useful.",
    "You have the powersource, now find some light producing device.",
    "You cannot put batteries into sometning that's not open.",
    "The more light, the better.",
    "Why not explore the darkest corners of the room?",
    "You really need a hint? Just take the key.",
    "You have the key and there is a lock in the door. Any ideas?",
    "Oh, c'mon!"]
]

location_pack = ["Now you are at the bed and the one-leg table.", "You have moved to the door and the TV set."]

stage = 0
location = 0
location_limit = "It's too far. You should get closer."
incorrect_input_msg_1 = "*** Please make a correct selection."
incorrect_input_msg_2 = "*** Please be more specific."

def delimit(printout):
    "Function for output highlighting"
    print("-"*72)
    print(printout)
    print("-"*72)

def delimit_strong(printout):
    "Function for output strong highlighting"
    print("\/"*36)
    print(printout)
    print("\/"*36)

def run():
    "A simple callout of the game"
    menu_input()

def menu_input():
    "Maintenance of the main menu input"

    global stage, location 
    main_menu()
    
    while True:
        try:
            uival = str(input("Your choise is: "))
            if uival == "1":
                new_game()
                break
            elif uival == "2":
                resume_game()
                break
            elif uival == "3":
                description()
            elif uival == "4":
                quit_game()
                break
            else:
                print(incorrect_input_msg_1)
        except Exception:
            print(incorrect_input_msg_1)

def main_menu():
    "Main menu of the game"
    global stage, location
    print("#"*72)
    for letter in "\t\tESCAPE THE ROOM\n":
        print(letter, end='')
        for _ in range(15000000):
            pass
    print("#"*72)
    print("\t\t[1] Start a new game")
    print("\t\t[2] Resume previous game")
    print("\t\t[3] Description")
    print("\t\t[4] Quit the game")
    print()

def new_game(): 
    "Starting a new game"
    global stage, location
    stage, location = 0, 0
    game_input()
    
def resume_game():
    "resuming the game in progress if any"
    global stage
    if stage == 0:
        print("*** No game in progress")
        menu_input()
    else:
        game_input()

def description():
    "Describes the main idea of the game"
    delimit("""To complete the game you need to explore the room and all objects. 
    \nThen you need to figure out which objects can be useful in tackling your objective.
    \nAnd the objective is to get out of the room.""")        
        
def quit_game():
    "Quit the game from the main menu"
    delimit("*** Thank you for playing the game...")
    
def game_input():
    "Maintenance of the game input"
    global stage, location

    stage_intro()
    game_actions()

    while True:
        try:
            uival = str(input("Your choise is: ")).lower()             
            if uival == "?":
                hlp()
            elif uival == "e":
                exit()
                break
            elif uival == "h":
                hint()
            elif uival == "la":
                look_around()
            elif uival == "m":
                move()
            elif uival == "i":
                inventory()
            elif uival[0] == "t":
                take(uival)
            elif uival[0] == "u":
                use(uival)
            elif uival[0] == "o":
                lcl_stage = stage
                open(uival)
                if uival == "o door" and lcl_stage == 9 and location == 1: 
                    lcl_stage = stage
                    break
            elif uival[0:2] == "lt":
                look_at(uival)
            else:
                print(incorrect_input_msg_2)
        except Exception:
            print(incorrect_input_msg_2)

def game_actions():
    "Available game actions" 
    print()
    print("Available game actions")
    print()
    print("\t\t[T] Take ...", "\t", "[LA] Look around")
    print("\t\t[U] Use ...", "\t", "[LT] Look at ...")
    print("\t\t[O] Open ...", "\t", "[H] Hint")
    print("\t\t[M] Move", "\t", "[?] Help")
    print("\t\t[I] Inventory", "\t", "[E] Exit to main menu")

def take(uival):
    "Take... action function"
    global stage
    if uival == "t remote":
        if stage == 0:
            delimit(f"{msg_pack[12][stage]}")
            stage +=1
            stage_intro()
        else: print(f"{msg_pack[12][stage]}")
    elif uival == "t flashlight":
            if stage == 3:
                delimit(f"{msg_pack[13][stage]}")
                stage += 1
                stage_intro()
            else: print(f"{msg_pack[13][stage]}")
    elif uival == "t key":
        if stage == 7:
            delimit(f"{msg_pack[14][stage]}")
            stage += 1
            stage_intro()
        else: print(f"{msg_pack[14][stage]}")
    else:
        print(incorrect_input_msg_2)

def use(uival):
    "Use... action function"
    global stage
    if uival == "u remote":
        print(f"{msg_pack[15][stage]}")
    elif uival == "u batteries":
        if stage == 5:
            delimit(f"{msg_pack[16][stage]}")
            stage += 1
            stage_intro()
        else: print(f"{msg_pack[16][stage]}")  
    elif uival == "u flashlight":
        if stage == 6:
            delimit(f"{msg_pack[17][stage]}")
            stage += 1
            stage_intro()
        else: print(f"{msg_pack[17][stage]}")     
    elif uival == "u key":
            if stage == 8 and location == 1:
                delimit(f"{msg_pack[18][stage]}")
                stage += 1
                stage_intro()
            elif stage == 8 and location == 0:
                print(location_limit)
            else: print(f"{msg_pack[18][stage]}")
    else:
        print(incorrect_input_msg_2)

def open(uival):
    "Open... action function"
    global stage
    if uival == "o remote":
        if stage == 2:
            delimit(f"{msg_pack[9][stage]}")
            stage += 1
            stage_intro()
        else: print(f"{msg_pack[9][stage]}")          
    elif uival == "o flashlight":
        if stage == 4:
            delimit(f"{msg_pack[10][stage]}")
            stage += 1
            stage_intro()
        else: print(f"{msg_pack[10][stage]}")        
    elif uival == "o door":
        if stage == 9 and location == 1:
            delimit(f"{msg_pack[11][stage]}")
            stage += 1
            stage_intro()
            stage = 0
            menu_input()
        elif stage == 9 and location == 0:
                print(location_limit)
        else: print(f"{msg_pack[11][stage]}")
    else:
        print(incorrect_input_msg_2)

def move():
    "Move action function"
    global location
    if location == 0:
        location = 1
        delimit(f"{location_pack[location]}")
    else:
        location = 0
        delimit(f"{location_pack[location]}")

def inventory():
    "Inventory list function"
    delimit(f"List of your inventory: {msg_pack[19][stage]}")

def look_around():
    "Look around action function"
    delimit(f"{msg_pack[1][stage]}\n\n{location_pack[location]}")

def look_at(uival):
    "Look at... action function"
    global stage
    if uival == "lt bed":
        print(f"{msg_pack[2][stage]}")
    elif uival == "lt table":
        print(f"{msg_pack[3][stage]}")
    elif uival == "lt tv":
        if stage == 1:
            delimit(f"{msg_pack[4][stage]}")
            stage += 1
            stage_intro()
        else: print(f"{msg_pack[4][stage]}")
    elif uival == "lt door":
        print(f"{msg_pack[5][stage]}")
    elif uival == "lt remote":
        print(f"{msg_pack[6][stage]}")
    elif uival == "lt flashlight":
        print(f"{msg_pack[7][stage]}")
    elif uival == "lt key":
        print(f"{msg_pack[8][stage]}")
    else:
        print(incorrect_input_msg_2)

def hint():
    "Hint action function"
    delimit(f"{msg_pack[20][stage]}")

def hlp():
    "Help action function"
    delimit("""Please select an action and an object for it if necessary. \
    \nHere are some sample dummy actions. Note that not all can be applied in the gameplay: \
    \n- to look around - LA\
    \n- to look at something - LT ceiling\
    \n- to take something - T bed\
    \n- to use something - U tv\
    \n- to open something - O table\
    \n- to move yourself, i.e. change your location - M\
    \n- to see your inventory - I
    \nFor input convenience use only short addressing such as described above, omit articles 'the' and 'a'\
    \nand type 'remote' instead of 'remote control' or 'tv' instead of 'tv set'.\
    \nThat means, for example, 'look around' instead of 'la' wont work.""")
    game_actions()

def exit():
    "Exit action function"
    menu_input()
    
def stage_intro():
    "New game-stage introduction"
    delimit_strong(f"{msg_pack[0][stage]}")
            
run()
```
